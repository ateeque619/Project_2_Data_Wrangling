```python
# Creating another copy of original DataFrame
testing_data = c_data.copy()
```

```python
# change the boolean value of holiday column with 0 and 1
testing_data['holiday']=testing_data['holiday'].replace({True: 1, False: 0})
```

```python
testing_data.describe()
```

```python
# check the outliers of temp column

sns.boxplot(y=testing_data['temp'])
plt.show()
```

```python
# handle the outlier of temp column by IQR and replace them with median value

Q1 = testing_data['temp'].quantile(0.25)
Q3 = testing_data['temp'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(Q1)
print(Q3)
print(IQR)
print(lower_bound)
print(upper_bound)

# Replace outliers with median
# numpy.clip() function is used to Clip (limit) the values in an array.

testing_data['temp'] = testing_data['temp'].clip(lower=lower_bound, upper=upper_bound)

# Check for outliers again
sns.boxplot(y=testing_data['temp'])
plt.show()
```

```python
# check the outliers of atemp column

sns.boxplot(y=testing_data['atemp'])
plt.show()
```

```python
# handle the outlier of atemp column by IQR and replace them with median value

Q1_a = testing_data['atemp'].quantile(0.25)
Q3_a = testing_data['atemp'].quantile(0.75)
IQR_a = Q3_a - Q1_a
lower_bound_a = Q1_a - 1.5 * IQR_a
upper_bound_a = Q3_a + 1.5 * IQR_a

print(Q1_a)
print(Q3_a)
print(IQR_a)
print(lower_bound_a)
print(upper_bound_a)

# Replace outliers with median
# numpy.clip() function is used to Clip (limit) the values in an array.

testing_data['atemp'] = testing_data['atemp'].clip(lower=lower_bound_a, upper=upper_bound_a)

# Check for outliers again
sns.boxplot(y=testing_data['atemp'])
plt.show()
```

```python
# check the outliers of hum column

sns.boxplot(y=testing_data['hum'])
plt.show()
```

```python
# check the outliers of windspeed column

sns.boxplot(y=testing_data['windspeed'])
plt.show()
```

```python
# handle the outlier of windspeed column by IQR and replace them with median value

Q1_w = testing_data['windspeed'].quantile(0.25)
Q3_w = testing_data['windspeed'].quantile(0.75)
IQR_w = Q3_w - Q1_w
lower_bound_w = Q1_w - 1.5 * IQR_w
upper_bound_w = Q3_w + 1.5 * IQR_w

print(Q1_w)
print(Q3_w)
print(IQR_w)
print(lower_bound_w)
print(upper_bound_w)

# Replace outliers with median
# numpy.clip() function is used to Clip (limit) the values in an array.

testing_data['windspeed'] = testing_data['windspeed'].clip(lower=lower_bound_w, upper=upper_bound_w)

# Check for outliers again
sns.boxplot(y=testing_data['windspeed'])
plt.show()
```

```python
#check the skewness

# Calculate the skewness for each numeric column
skewness = testing_data.skew()

# Print the skewness values
print("Skewness:")
print(skewness)
```

```python
# Calculate the correlation matrix
correlation_matrix = testing_data.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Season and year having the same values so it will not be showing the relationship
```

```python
# Create the heatmap
plt.subplots(figsize=(15,12))
sns.heatmap(testing_data.corr(), annot=True, cmap="coolwarm")

# Display the heatmap
plt.show()
```

```python
# Interpret the heatmap
# - The heatmap shows the correlation between each pair of columns in the dataset.
# - The values in the heatmap range from -1 to 1, where:
#   - 1 indicates a perfect positive correlation.
#   - 0 indicates no correlation.
#   - -1 indicates a perfect negative correlation.
# - The darker the color, the stronger the correlation.

# Insights:
# - There is a strong positive correlation between temp and atemp.
# - There is a strong negative correlation between temp and hum.
# - There is a weak positive correlation between temp and casual.
# - There is a weak negative correlation between atemp and casual.
# - There is a weak positive correlation between hum and casual.
# - There is a weak negative correlation between windspeed and casual.
```
